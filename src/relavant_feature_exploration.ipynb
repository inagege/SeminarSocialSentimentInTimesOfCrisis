```python
import pandas as pd
from tools.preprocessing_data import load_data
import matplotlib.pyplot as plt

def plot_development_over_time(data, target_column, grouped_period='W'):
    """
    Plot the development of a target column over time.
    @param data: The input data
    @param target_column: The target column to plot
    @param grouped_period: The period to group by (default: 'W' for week)
    
    """
    # Ensure the i_START column is in datetime format
    data['i_START'] = pd.to_datetime(data['i_START'])
    
    # Extract year and week from the datetime column
    data['year_' + grouped_period] = data['i_START'].dt.to_period(grouped_period)
    
    # Group by year_week and calculate the relevant statistics
    grouped_data = data.groupby('year_' + grouped_period).agg(
        mean_value=(target_column, 'mean'),
        std_value=(target_column, 'std'),
        median_value=(target_column, 'median'),
        q1_value=(target_column, lambda x: x.quantile(0.25)),  # First quartile (Q1)
        q3_value=(target_column, lambda x: x.quantile(0.75)),  # Third quartile (Q3)
        min_value=(target_column, 'min'),  # Minimum
        max_value=(target_column, 'max'),  # Maximum
        sample_count=(target_column, 'size')
    ).reset_index()
    
    # Convert 'year_week' back to datetime for plotting
    grouped_data['year_' + grouped_period] = grouped_data['year_' + grouped_period].dt.start_time
    
    # Plot the development of the target_column over time
    plt.figure(figsize=(24, 6))
    
    # Plot the mean values
    plt.plot(grouped_data['year_' + grouped_period], grouped_data['mean_value'], marker='o', linestyle='-', label='Mean Value')
    
    # Plot the shaded area representing the interquartile range (Q1 to Q3)
    plt.fill_between(grouped_data['year_' + grouped_period], 
                     grouped_data['q1_value'], 
                     grouped_data['q3_value'], 
                     color='blue', alpha=0.2, label='Interquartile Range (Q1-Q3)')
    
    # Plot the range between min and max as a whisker-like effect
    plt.fill_between(grouped_data['year_' + grouped_period], 
                     grouped_data['min_value'], 
                     grouped_data['max_value'], 
                     color='gray', alpha=0.1, label='Min-Max Range')

    # Plot the median as a separate line
    plt.plot(grouped_data['year_' + grouped_period], grouped_data['median_value'], marker='x', color='green', linestyle='--', label='Median')
    
    # Annotate each point with the sample count
    for i, row in grouped_data.iterrows():
        plt.text(row['year_' + grouped_period], row['mean_value'], f'{row["sample_count"]}', 
                 ha='center', va='bottom', fontsize=9)
    
    # Add labels and title
    plt.title('Development of ' + target_column + 'Over Time (Grouped by ' + grouped_period + ')')
    plt.xlabel(grouped_period)
    plt.ylabel('Value')
    plt.xticks(rotation=45)
    plt.grid(True)
    
    # Add a legend to the plot
    plt.legend()
    
    plt.show()
```

```python
plot_development_over_time(load_data(), 'F5A10_2', 'W')
```
